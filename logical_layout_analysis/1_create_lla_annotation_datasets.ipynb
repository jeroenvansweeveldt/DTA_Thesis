# Create datasets for logical layout analysis annotation

```python
import pandas as pd
```

_base_lla_datasets_ are generated using the _build_lla_dataset.py_ script, which combines the mapped filenames and the .hocr XHMTL data from the _ocr_output_ folder.

To allow for efficient annotation, these datasets should be enriched further with more extensive coordinate calculations, performed in this notebook.

```python
df = pd.read_csv("../data/base_lla_dataset_marescoe-david.csv")
print(df.dtypes)
df
```

Using the bounding box coordinates of the pages and the lines, we calculate the relative distances of each line from the page margins:

$$
\displaystyle
D = \frac{\text{Margin}}{\text{Page size in the chosen direction}}
$$

Where _D_ stands for the relative distance. We do this for all four different directions.

```python
df["Rel left"] = df["Line left"]/df["Page right"]
df["Rel right"] = df["Line right"]/df["Page right"]
df["Rel top"] = df["Line top"]/df["Page bottom"]
df["Rel bottom"] = df["Line bottom"]/df["Page bottom"]
df
```

```python
#Let's filter the dataframe on the header information
df_title = df[df['Line'].str.contains('The Marescoe-David Letters')]
df_title[['Filename','Line']].describe()
```

```python
#All the data without headers
df_body = df[~df['Line'].str.contains('The Marescoe-David Letters')]

df_body[['Filename','Line']].describe()
```

```python
cols = ["Rel left","Rel right","Rel top","Rel bottom"]
d = {col:{} for col in cols}
for col in cols:
    d[col]["mean"] = df_title[col].mean()
    d[col]["stdev"] = df_title[col].std()
    
print(d)
print(d["Rel left"]["mean"])
```

The following function checks for outliers in the relative distance values for all directions. If a value exceeds two standard deviations **above** the mean, it returns the difference as a positive value. If the value exceeds two standard deviations **below** the mean, it returns the difference as a negative value. If the value does not exceed the mean by two standard deviations, the function returns 0.

$$
\displaystyle
  D =
    \begin{cases}
      D - (\bar{D} + 2 \sigma_D) & \text{if } D > \bar{D} + 2 \sigma_D\\
      \bar{D} - 2 \sigma_D - D & \text{if } D < \bar{D} - 2 \sigma_D\\
      \text{else} & : 0
    \end{cases}       
$$

This is done to show which lines in the letters on the page show irregularities in distance to one another. This information is useful to identify where certain types of the letter structure are clustered (e.g. letter titles or IDs, end of alineas, etc.)

```python
def calculate_col(row, colname, avg, std):
    if row[colname] > avg+2*std:
        return row[colname] - (avg + 2 * std)
    elif row[colname] < avg-2*std:
        return avg - 2 * std - row[colname]
    else:
        return 0

def calculate_df(df, d):
    df["Corner left"] = df.apply(calculate_col, args=("Rel left",  d["Rel left"]["mean"], d["Rel left"]["stdev"]),axis=1)
    df["Corner right"] = df.apply(calculate_col, args=("Rel right", d["Rel right"]["mean"], d["Rel right"]["stdev"]),axis=1)
    df["Corner top"] = df.apply(calculate_col, args=("Rel top", d["Rel top"]["mean"], d["Rel top"]["stdev"]),axis=1)
    df["Corner bottom"] = df.apply(calculate_col, args=("Rel bottom", d["Rel bottom"]["mean"], d["Rel bottom"]["stdev"]),axis=1)

    corners = (df["Corner left"], df["Corner right"], df["Corner top"], df["Corner bottom"])
    df["Corner sum"] = sum(corners)

    return df["Corner left"], df["Corner right"], df["Corner top"], df["Corner bottom"], df["Corner sum"]
```

```python
calculate_df(df_body,d)
```

```python
df_body = df_body.sort_values(by="Corner sum", ascending=True)
df_body
```

For the next operation, we need to refer to Excel. In order to isolate footnotes more easily, we save our file to a .csv file and add another column in Excel, in which we will calculate the _previous line distance_ using Excel's formula functionalities.

We do this by checking whether the _"line bottom"_ column's current row value is **larger** than the preceding row's value. If **true**, take the _"line bottom"_ column's preceding row value, and **add** to that the **difference** between the current row value and the _"line top"_ column's preceding row value:

$$
\displaystyle
    \text{Previous line distance} =
    \begin{cases}
        LB_c > LB_p & : LB_p + (LB_c - LT_p) \\
        \text{else} & : 0
    \end{cases}
$$

$$
\displaystyle
    \text{Footer distance} = D_{\substack{\text{top}}} - \left(\frac{\text{previous line distance}}{\text{Page height}}\right)
$$
