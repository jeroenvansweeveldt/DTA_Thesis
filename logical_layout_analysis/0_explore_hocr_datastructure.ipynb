# Explore HOCR file structure

```python
import lxml
from lxml import etree
from xml.etree.ElementTree import Element
```

```python
tree = lxml.etree.parse("../ocr_output/hocr_marescoe-david/IMG_2022_07_05_12_07_39R.hocr")
print(lxml.etree.tostring(tree).decode())
```

```python
root = tree.getroot()
print(root)
print(root.tag)
print(root.attrib)
```

```python
for elem in root.findall('.//'):
    print(elem.attrib)
```

```python
for elem in root.findall('.//{http://www.w3.org/1999/xhtml}span[@class]'):
    print(elem.attrib, elem.text)
```
