Now, we will build the dataframe based on the info we need.

```python
def process_line(line):
    conf = []
    conf_val, wordlist = [], []
    children = line.getchildren()
    for child in children:
        wordlist.append(child.text)
        for token in child.get("title").split("; "):
            token = token.split(" ")
            conf.append(token)
        conf_val.append(token[1])
    conf_val = [int(i) for i in conf_val]
    line_id = line.get('id')
    coordinates = line.get('title').split(";")
    coordinates = coordinates[0].split(" ")
    coordinates = coordinates[1:]
    line = ' '.join(wordlist)
    confidence = conf_val
    avg = float(round(statistics.mean(conf_val), 2))
    stdev = 0.0
    if len(conf_val) > 1:
        stdev = round(statistics.stdev(conf_val), 2)

    return (line_id, coordinates, line, confidence, avg, stdev)
```

```python
def process_file(file,writer,f_num):
    tree = lxml.etree.parse(file)
    root = tree.getroot()

    p_info = root.find('.//{http://www.w3.org/1999/xhtml}div[@class="ocr_page"]')
    p_info = p_info.get("title")
    f_name = p_info.split('"')
    f_name = f_name[1].split("\\")
    f_name = f_name[-1]
    p_dim = p_info.split(";")
    p_dim = p_dim[1].split(" ")
    p_dim = p_dim[2:]

    
    lines = root.findall('.//{http://www.w3.org/1999/xhtml}span[@class="ocr_line"]')

    for l in lines:
        (line_id, coordinates, line, confidence, avg, stdev) = process_line(l)
        print(coordinates)
        print(line)
        print(confidence)
        row = {
            "Filename": f_name,
            "File number": f_num[f_name],
            "Line id": line_id,
            "Line":line,
            "Confidence":confidence,
            "Average":avg,
            "StDev": stdev
        }
        coor_pages = ["page left","page top","page right","page bottom"]
        for cp,title in zip(p_dim,coor_pages):
            row[title] = cp
        coor_titles = ["left","top","right","bottom"]
        for ct,title in zip(coordinates,coor_titles):
            row[title] = ct
        writer.writerow(row)
    return writer
```

```python
path = "scan_hocr/marescou_rescan"
files = os.listdir(path)

headers = {
        "Filename": "Filename",
        "File number": "File number",
        "page left": "Page left",
        "page top": "Page top",
        "page right": "Page right",
        "page bottom": "Page bottom",
        "Line id": "Line id",
        "Line": "Line",
        "Confidence": "Confidence",
        "Average":"Average",
        "StDev": "Standard deviation",
        "left": "Line left",
        "top": "Line top",
        "right": "Line right",
        "bottom": "Line bottom"
}
with open('marescou_rescan.csv', 'w') as file:
    writer = csv.DictWriter(file,fieldnames=headers)
    writer.writerow(headers)

    #Not possible to iterate as we normally would over the csv.reader object.
    #Stating 'f_num[f_map["filename"]] = [f_map["pagenr"]]' will result in
    #errors due to the object only recognising the headers.
    #Therefore, we first loop over the headers, then choose to continue.
    #To access the actual information we need, we need to name the rows
    #conform the variabel we defined in the loop.
    with open('mar_remap.csv') as csvfile:
        f_map = csv.reader(csvfile,delimiter=";")
        f_num = {}
        for i,row in enumerate(f_map):
            if i == 0:
                continue
            f_num[row[0]] = row[1]

    for f in files:
        f = os.path.join(path,f)
        print(f)
        if f == 'scan_hocr/marescou_rescan/.DS_Store':
            continue
        else:
            writer = process_file(f,writer,f_num)
```

```python
df["Rel left"] = df["Line left"]/df["Page right"]
df["Rel right"] = df["Line right"]/df["Page right"]
df["Rel top"] = df["Line top"]/df["Page bottom"]
df["Rel bottom"] = df["Line bottom"]/df["Page bottom"]
df
```

```python
#Let's filter the dataframe on the header information
df_title = df[df['Line'].str.contains('The Marescoe-David Letters')]
df_title[['Filename','Line']].describe()
```

```python
#All the data without headers
df_body = df[~df['Line'].str.contains('The Marescoe-David Letters')]

df_body[['Filename','Line']].describe()
```

```python
cols = ["Rel left","Rel right","Rel top","Rel bottom"]
d = {col:{} for col in cols}
for col in cols:
    d[col]["mean"] = df_title[col].mean()
    d[col]["stdev"] = df_title[col].std()
    
print(d)
print(d["Rel left"]["mean"])
```

```python
# We calculate the corners of the margins, line per line, 
# so we will be able to see through the sum of all corners per line
# which lines shows irregularities.
# This is to isolate e.g. letter IDs in the margins,
# or where alineas end.

def calculate_col(row, colname, avg, std):
    if row[colname] > avg+2*std:
        return row[colname] - (avg + 2 * std)
    elif row[colname] < avg-2*std:
        return avg - 2 * std - row[colname]
    else:
        return 0

def calculate_df(df, d):
    df["Corner left"] = df.apply(calculate_col, args=("Rel left",  d["Rel left"]["mean"], d["Rel left"]["stdev"]),axis=1)
    df["Corner right"] = df.apply(calculate_col, args=("Rel right", d["Rel right"]["mean"], d["Rel right"]["stdev"]),axis=1)
    df["Corner top"] = df.apply(calculate_col, args=("Rel top", d["Rel top"]["mean"], d["Rel top"]["stdev"]),axis=1)
    df["Corner bottom"] = df.apply(calculate_col, args=("Rel bottom", d["Rel bottom"]["mean"], d["Rel bottom"]["stdev"]),axis=1)

    corners = (df["Corner left"], df["Corner right"], df["Corner top"], df["Corner bottom"])
    df["Corner sum"] = sum(corners)

    return df["Corner left"], df["Corner right"], df["Corner top"], df["Corner bottom"], df["Corner sum"]
```
